Load the data

```python
import pandas as pd
import numpy as np

data = pd.io.stata.read_stata('data/shor+mccarty+1993-2016+individual+legislator+data+May+2018+release+(Updated+July+2018).dta')

print(data.shape)
```

Get columns in data

```python
columns = data.columns

print(columns)
```

Get data types per column

```python
print(type(data.dtypes))
print(data.name.dtype)
print(data.name.value_counts())
```

Check for nulls

```python
print(data.isnull().sum())
```

```python
print(len(data.st.unique()))
print(data.party.unique())
print(data.party.value_counts())
print(data.name.value_counts())
```

```python
print(len(data.name.unique()))

```

```python
columnDatatypes = data.dtypes

for i, v in columnDatatypes.iteritems():
    print('==================')
    print(f'{i}: {v}\n')
    
    if data[i].dtype.name in ['object', 'category']:
        print(data[i].value_counts())

    else:
        print(f'min: {data[i].min()}')
        print(f'median: {data[i].mean()}')
        print(f'max: {data[i].max()}')
        print(f'std: {data[i].std()}')
    
    print(f'Nulls: {data[i].isnull().sum()}')
    print('------------------')
```
